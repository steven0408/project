```python
# 建立棋盤
def BuildBoard(BoardSize):
    return np.zeros((BoardSize, BoardSize))
    
# 列印棋盤
def PrintBoard(Board):
    for i in Board:
        print(np.where(i == 1, "●", np.where(i == -1, "○", " ")))
    print("\n")
```

```python
# 獲取可以下棋位置
# 回傳 list，eq. [[2,3],[3,4]]
def GetValid(Board):
    return [[i,j] for i in range(len(Board)) for j in range(len(Board)) if Board[i,j] == 0]

# 依據 action 判斷連續
# 目前只適合 8 * 8 棋盤
def IsContinuous(Board, action):
    row, col = action

    # 橫列
    s = np.where(Board[row] == Board[row,col])[0]
    if len(s) == 5:
        return True if s[0] + 4 == s[4] else False
    # 直行
    s = np.where(Board[:, col] == Board[row,col])[0]
    if len(s) == 5:
        return True if s[0] + 4 == s[4] else False

    # 正斜
    t = [Board[i,j] for i in range(len(Board)) for j in range(len(Board)) if i + j == row + col]
    s = np.where(t == Board[row, col])[0]
    if len(s) == 5:
        return True if s[0] + 4 == s[4] else False

    # 反斜
    t = [Board[i,j] for i in range(len(Board)) for j in range(len(Board)) if i - j == row - col]
    s = np.where(t == Board[row, col])[0]
    if len(s) == 5:
        return True if s[0] + 4 == s[4] else False
    
    return False
```

```python
import random
def Player1_move(Board, action, chessman):
    return random.choice(GetValid(Board))
def Player1_choose(Board, action, chessman):
    return random.choice([-1, 0, 1])
def Player2_move(Board, action, chessman):
    return random.choice(GetValid(Board))
def Player2_choose(Board, action, chessman):
    return random.choice([-1, 0, 1])
```

```python
Board = BuildBoard(8)
Player1_score = 0
Player2_score = 0
draw = 0
Times = 10   # 執行次數
swap = True # 黑白棋子決定與否 ( True: 尚未決定)

# 設置 player

# 主程式
for i in range(Times):
    action = []
    # 決定誰假先手?
    CurrentPlayer = random.choice(["Player1", "Player2"])
    chessman = 1    # 先手黑棋
    step = 0

    while GetValid(Board) != []:
        if CurrentPlayer == "Player1":
            action = Player1_move(Board, action, chessman)
            if step < 6 and step != 3 and swap:
                if action in GetValid(Board):
                    row, col = action
                    Board[row, col] = chessman
                    chessman = -chessman
                    step += 1
                    if step == 2 or step == 5:
                        CurrentPlayer = 'Player2'
                else:
                    print('Error')
                    break
            elif step == 3 and swap:
                step += 1
                option = Player1_choose(Board, action, chessman)
                if option != 0:
                    swap = False
                    if option == 1:
                        CurrentPlayer = 'Player2'

            elif step == 6 and swap:
                while option == 0:
                    option = Player1_choose(Board, action, chessman)
                step += 1
                swap = False
                if option == 1:
                    CurrentPlayer = 'Player2'
                    
            else:
                if action in GetValid(Board):
                    row, col = action
                    Board[row, col] = chessman
                    if IsContinuous(Board, action):
                        Player1_score += 1
                        print("Player1 win")
                        break
                    else:
                        CurrentPlayer = "Player2"
                        chessman = -chessman
                else:
                    print("Error")
                    break
        else:
            action = Player2_move(Board, action, chessman)
            if step < 6 and step != 3 and swap:
                if action in GetValid(Board):
                    row, col = action
                    Board[row, col] = chessman
                    chessman = -chessman
                    step += 1
                    if step == 2 or step == 5:
                        CurrentPlayer = 'Player1'
                else:
                    print('Error')
                    break
            elif step == 3 and swap:
                step += 1
                option = Player2_choose(Board, action, chessman)
                if option != 0:
                    swap = False
                    if option == 1:
                        CurrentPlayer = 'Player1'
                        
            elif step == 6 and swap:
                while option == 0:
                    option = Player2_choose(Board, action, chessman)
                step += 1
                swap = False
                if option == 1:
                    CurrentPlayer = 'Player1'
            else:
                if action in GetValid(Board):
                    row, col = action
                    Board[row, col] = chessman
                    if IsContinuous(Board, action):
                        print("Player2 win")
                        Player2_score += 1
                        break
                    else:
                        CurrentPlayer = "Player1"
                        chessman = -chessman
                else:
                    print("Error")
                    break

    PrintBoard(Board)
    Board = BuildBoard(8)            
    
# 輸出結果
print("Player1 的勝率 : {}\nPlayer2 的勝率 : {}\n平手 : {}".format(Player1_score/Times, Player2_score/Times, (Times-Player1_score-Player2_score)/Times))
```
